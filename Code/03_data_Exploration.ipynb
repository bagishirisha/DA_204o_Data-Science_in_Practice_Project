## STEP 3: Data Exploration

### Import necessary libraries

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

### 3.1 Load datasets from file paths

```python
data_path = '..\Dataset\data_cleaned.csv'
genre_data_path = '..\Dataset\genre_data_cleaned.csv'

# Check if files exist and load them
if os.path.exists(data_path) and os.path.exists(genre_data_path):
    data = pd.read_csv(data_path)
    genre_data = pd.read_csv(genre_data_path)
    print("Info: Data and genre data successfully loaded.")
else:
    print("Attention: One or both files are not found in the specified directory.")
```

### 3.2 Observations based on popularity

#### 3.2.1 Least popular songs and the similarity in their features

```python
#Sort the dataset based on popularity and observe the least 5 popular tracks
least_popular = data[(data['popularity'] <= 10) & (data['popularity'] >= 0)].sort_values('popularity', ascending=True)

least_popular_table = least_popular[['year', 'mode_category', 
                                     'danceability_category', 'energy_category', 'valence_category', 
                                     'explicit_category','release_decade','tempo_category','key_category',
                                     'acousticness_category','liveness_category', 'instrumentalness_category','speechiness_category']]

top_least_popular = least_popular[['popularity','year', 'name', 'artists', 'mode_category', 
                                   'danceability_category', 'energy_category', 'valence_category', 
                                   'energy_category']]

#find common feature values among the least popular songs (between 0 and 10 popularity score)
common_values = {}
for column in least_popular_table.columns:
    common_values[column] = least_popular_table[column].mode().values[0]  # Finding the most common value



print('Top 5 least popular tracks:')
display(top_least_popular.head(5) )

print('\n')
print('Common features among the least popular tracks')
display(common_values)
```

##### **Conclusions based on Least popular songs**
- The top 5 least popular songs as printed in above cell, and they all have popularity score of 0 .
- The least popular songs (having popularity score between 0 to 10) have common features like very low energy,low danceability , low valence- emotionally negative or neutrals, low tempo , low key-typically quieter tonal center,Moderate live presence,mostly instrumental and from release decade 1940's.

#### 3.2.2 Most popular songs and the similarity in their features

```python
#Sort the dataset based on popularity and observe the top 5 popular tracks
most_popular = data[(data['popularity'] <= 100) & (data['popularity'] >= 90)].sort_values('popularity', ascending=False)

most_popular_table = most_popular[['year', 'mode_category', 
                                     'danceability_category', 'energy_category', 'valence_category', 
                                     'explicit_category','release_decade','tempo_category','key_category',
                                     'acousticness_category','liveness_category', 'instrumentalness_category','speechiness_category']]

top_most_popular = most_popular[['popularity','year', 'name', 'artists', 'mode_category', 
                                   'danceability_category', 'energy_category', 'valence_category', 
                                   'energy_category']]
common_values_most_popular = {}
for column in most_popular_table.columns:
    common_values_most_popular[column] = most_popular_table[column].mode().values[0]  # Most common value

print('Top 5 most popular tracks:')
display(top_most_popular.head(5))
        
print('\n')
print('Common features among the most popular tracks')
display(common_values_most_popular)
```

##### **Conclusions based on Most popular songs**
- The song 'Dakiti' is the most popular song with popularity score of 100.It has high energy , neutral emotions and high danceability.
- The most popular songs (having popularity score between 90 to 100) have common features like moderate energy,high danceability , low valence- emotionally negative or neutrals, high tempo,very low live presence,no instrumental elements, primarily vocal driven and from release decade 2020's.

#### 3.3.2 Distribution of Categorical Features

```python
def adjust_xtick_labels(ax, labels, max_char=20):
    new_labels = []
    for label in labels:
        if len(label) > max_char:
            words = label.split()
            new_label = ''
            current_line = ''
            for word in words:
                if len(current_line) + len(word) + 1 <= max_char:
                    current_line += word + ' '
                else:
                    new_label += current_line.strip() + '\n'
                    current_line = word + ' '
            new_label += current_line.strip()
            new_labels.append(new_label)
        else:
            new_labels.append(label)

    ax.set_xticklabels(new_labels, rotation=45, ha='right')
```

```python
plt.figure(figsize=(20, 30))

data_features = ['tempo_category', 'mode_category', 'explicit_category',
                 'popularity_category', 'danceability_category', 'energy_category',
                 'valence_category', 'speechiness_category', 'acousticness_category',
                 'instrumentalness_category', 'liveness_category', 'duration_ms_category']

color_palette = sns.color_palette("Set1", len(data_features))

for i, feature in enumerate(data_features, 1):
    ax = plt.subplot(6, 2, i)  
    counts = data[feature].value_counts()
    colors = color_palette[:len(counts)]  
    counts.plot(kind='bar', color=colors, edgecolor='black', alpha=0.8)
    plt.title(f'Distribution of {feature}', fontsize=14)
    plt.xlabel(feature, fontsize=12)
    plt.ylabel('Frequency', fontsize=12)
    adjust_xtick_labels(ax, counts.index, max_char=25)

    plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.subplots_adjust(left=0.1, right=0.9, top=2.5, bottom=1, hspace=0.5, wspace=0.3)
plt.show()
```

##### **Conclusion**

This series of bar charts provides a breakdown of various categorized features in the dataset, displaying the distribution of songs across different features. Here’s an analysis of each chart:

- **Tempo Category**: The songs are evenly distributed across different tempo ranges, indicating a balanced representation of slow, moderate, and fast tracks.

- **Energy Category**: The energy levels are balanced across categories, indicating that the dataset includes both mellow and energetic songs, offering variety in terms of intensity.

- **Valence Category**: Songs are well distributed across different mood levels, from low to high positivity, reflecting a range of emotional content from somber to cheerful.

- **Speechiness Category**: Most songs fall into moderate to low speech content, meaning the tracks are likely more musical rather than speech-dominant audio, like podcasts or spoken-word recordings.

- **Acousticness Category**: There is an even distribution of songs across acousticness levels, from highly acoustic to more electronically produced sounds, allowing for a variety of production styles.

- **Instrumentalness Category**: Songs are fairly evenly split between instrumental and vocal tracks, with slightly more tracks containing vocals. This suggests that there’s a slight preference for tracks with vocals.

### 3.4 Bivariate Analysis - Numerical vs. Numerical Relationships

#### 3.4.1 Correlation Analysis - Heatmap of Numerical Feature Correlations

```python
# Plot a heatmap to identify correlations among numerical variables
numerical_data = data.select_dtypes(include=['float64', 'int64'])
 
# Calculating the correlation matrix
correlation_matrix = numerical_data.corr()
 
# Plotting the heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True)
plt.title('Correlation Matrix of Features')
plt.show()
```

##### **Conclusion**

The correlation heatmap is generated for all numerical features. Below are some key observations.

- **Valence and Danceability**: There is a moderate positive correlation between valence and danceability (0.56). This suggests that songs with a higher valence (happier or more positive mood) tend to also be more danceable. This makes sense, as upbeat, positive songs are often more suitable for dancing.

- **Year and Popularity**: There is a strong correlation between year and popularit (0.86) indicating that newer songs are much more likely to be popular in this dataset.

- **Energy and Loudness_scaled**: Energy and loudness have a strong positive correlation (0.78), which is expected as louder tracks often convey a higher energy level.

- **Acousticness and Energy**: Acousticness is negatively correlated with energy (-0.75), indicating that acoustic songs tend to have lower energy levels.

- **Acousticness and Year**: There is a moderate negative correlation between acousticness and year (-0.61). This suggests that newer songs tend to be less acoustic, possibly due to the increased use of electronic production techniques in recent music.

#### 3.4.2 Heatmap - Audio Features by Popularity Category

```python
# Grouping the data by 'popularity_category' and calculating the mean for each audio feature
avg_features_by_popularity = data.groupby('popularity_category')[audio_features_data].mean()
plt.figure(figsize=(12, 8))

# Creating a heatmap to display the average audio features by popularity category
sns.heatmap(avg_features_by_popularity, annot=True, cmap='magma', fmt=".2f")
plt.title("Average Audio Features by Popularity Category")
plt.ylabel("Popularity Category")
plt.xlabel("Audio Features")
plt.show()
```

##### **Conclusion**

This heatmap shows the average values of various audio features across different popularity categories. Here are some insights based on the observed patterns:

- **Danceability**: Tracks with higher popularity have a slightly higher average danceability (0.59) compared to tracks with very low popularity (0.50). This suggests that more danceable songs tend to be more popular.
- **Energy**: There is a noticeable difference in energy across popularity categories. Songs with high popularity have an average energy of 0.62, while very low popularity songs have a significantly lower energy level (0.28). This indicates that higher-energy tracks are more likely to be popular.
- **Acousticness**: Acousticness is lower for more popular songs (0.28 in high popularity vs. 0.84 in very low popularity), suggesting that more popular songs are less acoustic and possibly more electronically produced or digitally enhanced.
- **Tempo**: There’s a notable difference in tempo across popularity categories. Songs with high popularity have a tempo of around 120.59 BPM, while very low popularity songs have a slower tempo (109.87 BPM). This suggests that slightly faster-paced songs are generally more popular.
-**Loudness** (scaled): Loudness values are higher for more popular songs (0.81 in high popularity vs. 0.70 in very low popularity). This suggests that louder, more dynamic tracks are likely to be more popular, as they may feel more energetic and engaging.

#### 3.4.4 Trend analysis

- **Average Popularity of Songs Over the Years** : Song popularity has seen steady growth over time, with significant boosts during periods of technological advancement in music distribution and accessibility.i.e. between 1980s to 2020s.

- **Count of Songs by Release Decade** : The music industry has sustained high levels of song production from the 1950s onwards, with significant growth in earlier years as recording technology and distribution developed.In 2020s there is a noticeable drop in song count, likely due to incomplete data for this decade or the limited number of years covered within it so far.
